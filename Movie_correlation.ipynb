```python
import pandas as pd
import numpy as np
import seaborn as sns
import chardet


import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv('movies.csv',encoding= 'unicode_escape')
```

```python
df.head()
```

```python
#1.provjera da li ima missing data preko LOOP

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
   
```

```python
#provjeriti data types
print(df.dtypes)
```

```python
# promjena data type (rješavamo se nule iza decimalne točke)

df['budget']=df['budget'].astype('int')

df['gross']=df['gross'].astype('int')

```

```python
#vizualizacija korelacije
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#provjeravamo povezanost kompanije i zarade

#kompanije su str i morami ih transformirati

df_numerized =df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
      

```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_matrix= df_numerized.corr()
corr_pair= correlation_matrix.unstack()

corr_pair
```

```python
sorted_pairs = corr_pair.sort_values()
sorted_pairs
```

```python
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
